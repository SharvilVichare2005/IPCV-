```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import cv2
Frame = cv2.imread('/content/single-fresh-red-strawberry-on-table-green-background-food-fruit-sweet-macro-juicy-plant-image-photo.jpg')
plt.imshow(Frame)
```

```python
def Canny_detector(image, low_threshold=50, high_threshold=150):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    canny = cv2.Canny(blurred, low_threshold, high_threshold)
    return canny
canny_img = Canny_detector(Frame)
```

```python
plt.imshow(canny_img , cmap = "grey")
```

```python
def Sobel_detector(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    sobel = np.sqrt(sobelx**2 + sobely**2)
    sobel = cv2.normalize(sobel, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    return sobel
sobel_img = Sobel_detector(Frame)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(canny_img, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_img, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```
